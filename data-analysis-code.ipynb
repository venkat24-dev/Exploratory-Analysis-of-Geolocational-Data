```python
import pandas as pd

# Loading the dataset
geo_data = pd.read_csv(r"geolocational_data.csv")
data = pd.read_csv(r"food_coded.csv") 

# Display the first few rows
data.head()
```

```python
# Remove duplicates
data_no_duplicates = data_ffill.drop_duplicates()
```

```python
# Check available columns first
print(data_no_duplicates.columns.tolist())
```

```python

# Select specific columns for analysis (only if they exist)
columns_to_use = ['cook', 'eating_out', 'employment', 'ethnic_food']
cleaned_data = data_no_duplicates[columns_to_use].copy()
```

```python

# Display first few rows
cleaned_data.head()
```

```python
# In [8]
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(data=cleaned_data)
plt.title('Distribution of Features')
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# In [9]
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia_values = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(cleaned_data.select_dtypes(include='number'))
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia_values, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

```python
# In [10]
data = {
    'Latitude': [12.9228, 12.9686, 12.9534, 12.9434, 12.9735, 12.9956, 12.8975, 12.9640],
    'Longitude': [77.5921, 77.6066, 77.5940, 77.6853, 77.1026, 77.8975, 77.2454, 77.1200],
    'Restaurant_Count': [10, 7, 12, 5, 1, 8, 6, 20],
    'Gym_Count': [2, 3, 4, 3, 8, 9, 6, 4],
    'School_Count': [3, 4, 5, 5, 7, 12, 16, 13],
    'Hospital_Count': [3, 2, 4, 12, 14, 18, 20, 1]
}

df = pd.DataFrame(data)
df.to_csv('geolocational_data.csv', index=False)
df.head()
```

```python
# In [11]
# Select Latitude, Longitude, and one feature for clustering
feature = 'Restaurant_Count'
df2 = df[['Latitude', 'Longitude', feature]]

# Elbow Method for this feature
inertia_values = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df2[[feature]])
    inertia_values.append(kmeans.inertia_)

plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title(f'Elbow Method for {feature}')
plt.show()
```

```python
# In [12]
import folium
from sklearn.cluster import KMeans

# Fit KMeans clustering (assuming K=4)
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[['Latitude', 'Longitude']])

# Create a Folium map
center_lat, center_lon = df['Latitude'].mean(), df['Longitude'].mean()
m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# Assign colors to clusters
colors = ['red', 'blue', 'green', 'purple']

for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5,
        color=colors[row['Cluster']],
        fill=True,
        fill_color=colors[row['Cluster']],
        popup=f"Cluster: {row['Cluster']}"
    ).add_to(m)

# Save interactive map
m.save('clustered_locations_map.html')
m
```
